```python
%matplotlib inline
import matplotlib.pyplot as plt
from i3alerts.effective_area import load_aeff
```

```python
selections = ["alerts_v1", "alerts_v2", "ps_tracks"]
declination_deg = 0.
```

```python
plt.figure()

for selection in selections:
    aeff = load_aeff(declination_deg=declination_deg, selection=selection)
    for a in aeff:
        plt.plot(a["E_TeV"], a["A_eff"], label=selection)
        
plt.yscale("log")
plt.xscale("log")
plt.xlabel(r"$E_{\nu}$ [TeV]")
plt.ylabel(r"$A_{eff}$ [m]")
plt.legend()
```
